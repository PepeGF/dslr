```python
"""Displays pair plot or scatter plot matrix for given data."""

from pathlib import Path

import pandas as pd
import seaborn as sns

from aux_functions import read_data
```

```python
dataset_path: Path = "data/dataset_train.csv"
```

```python
df: pd.DataFrame = read_data(dataset_path)
```

```python
cols_to_plot: list[str] = (
    df.select_dtypes(include="number")
    .columns.drop(
        "Index",
        errors="raise",
    )
    .to_list()
)
house_colors: dict[str, str] = {
    "Gryffindor": "red",
    "Hufflepuff": "orange",
    "Ravenclaw": "blue",
    "Slytherin": "green",
}
```

```python
pair_grid: sns.PairGrid = sns.pairplot(
    df,
    vars=cols_to_plot,
    hue="Hogwarts House",
    palette=house_colors,
    diag_kind="hist",
)
pair_grid.figure.suptitle("Hogwarts House Pair Plot", y=1.02)
```

```python
output_path: Path = "data/pair_plot.png"
pair_grid.savefig(output_path)
```
